```python
dfs = [apple, amazon,facebook, google, microsoft, nvidia, tesla]
```

```python
# add two new columns that gives us the 50 day moving average and 200 day moving average
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
apple.head(50)
```

```python
 # add a column which show the previous day's closing price 

for df in dfs:
    df['previous day close price'] = df.Close.shift(1)
```

```python
tesla.head(10)
```

```python
# create a cloumn that shows the change in price at the end of the day

for df in dfs:
    df['change in price'] = df['Close'] - df['previous day close price']
```

```python
apple.head(10)
```

```python
# add column which shows percentage change in price

for df in dfs:
    df['percentage change in price'] = df['Close'].pct_change()
```

```python
# add a previous day volumne column

for df in dfs:
    df['previous day volume'] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
# change in vloume
for df in dfs:
    df['change in volume'] = df['Volume'] - df['previous day volume']
```

```python
# percentage change in volume
for df in dfs:
    df['percent change in volume'] = df.Volume.pct_change()
```
